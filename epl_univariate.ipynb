```python
#step1
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
%matplotlib inline 
```

```python
#step2
dataframe = pd.read_csv('EPL_20_21.csv')
goals_mins_player = dataframe.loc[:,['Name','Goals','Mins']]
x_train = goals_mins_player['Mins']
y_train = goals_mins_player['Goals']
```

```python
#step3 
m = len(x_train)
```

```python
#step4 
plt.scatter(x_train, y_train, marker = 'x', c = 'r')
plt.title('EPL data')
plt.xlabel('Minutes Played')
plt.ylabel('Goals Scored')
plt.show()
```

```python
#step5 
w = 0.004
b = 0.004
def compute_model_output(x,w,b):
    m = len(x)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
tmp_f_wb = compute_model_output(x_train, w, b)

#step 6
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.title('EPL data')
plt.xlabel('Mins played')
plt.ylabel('Goals scored')
plt.legend()
plt.show()
```
